```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### 1) transaction_v2 전처리 하기 (user 별로 정리하기)

```python
transaction_v2 = pd.read_csv("D:/Users/Kwon/Desktop/DATABASE/kkbox-churn-prediction-challenge/first/transactions_v2.csv/data/churn_comp_refresh/transactions_v2.csv")
transaction_v1 = pd.read_csv("D:/Users/Kwon/Desktop/DATABASE/kkbox-churn-prediction-challenge/transactions.csv/transactions.csv")

tv2_out1 = transaction_v2[transaction_v2["plan_list_price"]==0].index.values
lv2_1 = tv2_out1.tolist()
tv2_out2 = transaction_v2[transaction_v2["plan_list_price"] != transaction_v2["actual_amount_paid"]].index.values
lv2_2 = tv2_out2.tolist()
tv2_out3 = transaction_v2[transaction_v2["payment_plan_days"]== 0].index.values
lv2_3 = tv2_out3.tolist()

set2_1 = set(lv2_1)
set2_2 = set(lv2_2)
set2_3 = set(lv2_3)
set_2_all = set2_1.union(set2_2,set2_3)
trans_v2 = transaction_v2.drop(set_2_all,axis = 0)

trans_v2_id_count = trans_v2["msno"].value_counts()
trans_v2_count = pd.DataFrame(trans_v2_id_count, columns = ["msno"])
trans_v2_count = trans_v2_count.sort_index()
trans_v2_count = trans_v2_count.rename(columns = {"msno":"count"})

trans_v2_group = trans_v2.groupby("msno").aggregate({"payment_plan_days" : ["mean"],"plan_list_price" : ["mean","sum"]})
trans_v2_all = pd.merge(left = trans_v2_group,right = trans_v2_count,how = "inner",left_index = True, right_index = True)
```

### 1)_2 transaction_v1 전처리 하기 (user 별로 정리하기)

```python
tv1_out1 = transaction_v1[transaction_v1["plan_list_price"]==0].index.values
lv1_1 = tv1_out1.tolist()
tv1_out2 = transaction_v1[transaction_v1["plan_list_price"] != transaction_v1["actual_amount_paid"]].index.values
lv1_2 = tv1_out2.tolist()
tv1_out3 = transaction_v1[transaction_v1["payment_plan_days"]== 0].index.values
lv1_3 = tv1_out3.tolist()

set1_1 = set(lv1_1)
set1_2 = set(lv1_2)
set1_3 = set(lv1_3)
set_1_all = set1_1.union(set1_2,set1_3)
trans_v1 = transaction_v1.drop(set_1_all,axis = 0)

trans_v1_id_count = trans_v1["msno"].value_counts()
trans_v1_count = pd.DataFrame(trans_v1_id_count, columns = ["msno"])

trans_v1_count = trans_v1_count.sort_index()
trans_v1_count = trans_v1_count.rename(columns = {"msno":"count"})

trans_v1_group = trans_v1.groupby("msno").aggregate({"payment_plan_days" : ["mean"],"plan_list_price" : ["mean","sum"]})
trans_v1_all = pd.merge(left = trans_v1_group,right = trans_v1_count,how = "inner",left_index = True, right_index = True)



```

### 2) transaction_v1 v2 합치기

```python
trans_all = pd.merge(left = trans_v1_all,right = trans_v2_all,how = "inner",left_index = True, right_index = True)


trans_all = trans_all.fillna(0)

trans_all = trans_all.sort_index()
trans_all1= trans_all[trans_all.columns[0]]
trans_all5= trans_all[trans_all.columns[4]]

trans_all2= trans_all[trans_all.columns[1]]
trans_all6= trans_all[trans_all.columns[5]]

trans_all3= trans_all[trans_all.columns[2]]
trans_all7= trans_all[trans_all.columns[6]]

trans_all4= trans_all[trans_all.columns[3]]
trans_all8= trans_all[trans_all.columns[7]]

trans_1 = pd.DataFrame((trans_all1 + trans_all5)/2,columns = ["payment_plan_days_mean"] )
trans_2 = pd.DataFrame((trans_all2 + trans_all6)/2,columns = ["plan_list_price_mean"])
trans_3 = pd.DataFrame(trans_all3 + trans_all7,columns = ["plan_list_price_sum"])
trans_4 = pd.DataFrame(trans_all4 + trans_all8,columns = ["count_sum"])

trans_12 = pd.merge(left = trans_1,right = trans_2,how = "inner",left_index = True, right_index = True)
trans_34 = pd.merge(left = trans_3,right = trans_4,how = "inner",left_index = True, right_index = True)
t_all = pd.merge(left = trans_12,right = trans_34,how = "inner",left_index = True, right_index = True)
t_all
```
