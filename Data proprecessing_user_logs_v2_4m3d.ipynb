### 한번에 한번 aggregate 할때는 => 
    data.groupby(by = "기준칼럼").계산칼럼.agg(...)
### 한번에 여러개의 칼럼 aggragate 할떄는 =>
    data.groupby(by = "기준칼럼").agg({'계산칼럼' : 계산식 , '계산칼럼2' : 계산식2})

```python
ul_v2_num = user_logs_v2.drop(['date','num_unq','total_secs'],axis = 1)

ul_v2_num_mean = ul_v2_num.groupby(by = "msno").aggregate('mean')
ul_v2_num_mean
```

```python
len(ul_v2_num_mean)
### transaction 에서 outlier 제거한 유저들도 포함되어 있어서 90만 이 아니라 100만 명 정도 있는 거!
```

```python
ul_v2_s = user_logs_v2.drop(["num_25","num_50","num_75","num_985","num_100","date"],axis = 1)
ul_v2_sum = ul_v2_s.groupby(by = "msno").aggregate("sum")
ul_v2_sum
```

```python
len(ul_v2_sum)
```

```python
ul_v2_sum_sort = ul_v2_sum.sort_values(by= "num_unq")
ul_v2_sum_sort
```

```python
ul_v2_sum_unq1 = ul_v2_sum_sort[ul_v2_sum_sort["num_unq"]==1]
len(ul_v2_sum_unq1)
```

```python
ul_v2_sum_unq10 = ul_v2_sum_sort[ul_v2_sum_sort["num_unq"] < 10]
len(ul_v2_sum_unq10)
```

```python
ul_v2_sum_count = ul_v2_sum_sort.drop(["total_secs"],axis =1).apply(pd.value_counts)
ul_v2_sum_count
```
