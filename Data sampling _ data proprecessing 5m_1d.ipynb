```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
transaction_all = pd.read_csv("transcation_all.csv")
```

```python
transaction_all = transaction_all.sort_values(by = "count_sum")

quantile_data = transaction_all[["msno","count_sum"]]
quantile_data_2_index = transaction_all[transaction_all["count_sum"] == 1 ].index.values
quantile_data_2 = transaction_all.drop(quantile_data_2_index,axis = 0)
quantile_data_3 = transaction_all[transaction_all["count_sum"] == 1]

#### array 1 = all , array 2 = array1.drop(count_sum ==1) , array_3 = array1[count_sum == 1]

quantile_list = quantile_data.values.tolist()
quantile_list_2 = quantile_data_2.values.tolist()
quantile_list_3 = quantile_data_3.values.tolist()

```

```python
len(quantile_list),len(quantile_list_2),len(quantile_list_3)
```

```python
### quantile_array_2 에서 3개 그룹으로 나누기
### quantile_array_3 = 1개 그룹

quantile_2_1 = round(len(quantile_list_2) * (1/3))
quantile_2_2 = round(len(quantile_list_2) * (2/3))
quantile_2_3 = round(len(quantile_list_2) * (3/3))

group_2_1 = []
group_2_2 = [] 
group_2_3 = []
group_2_4 = quantile_list_3
 
for num in range(len(quantile_list_2)):
    if num < quantile_2_1 :
        group_2_1.append(quantile_list_2[num])
    elif num < quantile_2_2 :
        group_2_2.append(quantile_list_2[num])
    else : 
        group_2_3.append(quantile_list_2[num])

```
